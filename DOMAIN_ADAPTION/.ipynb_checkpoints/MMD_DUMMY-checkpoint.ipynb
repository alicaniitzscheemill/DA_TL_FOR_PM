```python
def create_source_data(n):

    mean_class_1_domain_1  = [0, -10, 0, 0, -10, 0, 0, -10, 0]
    cov_class_1_domain_1  = np.diag([1,2,6,4,2,1,8,8,6])
    X_data_class_1_domain_1  = np.random.multivariate_normal(mean_class_1_domain_1, cov_class_1_domain_1, n)
    y_data_class_1_domain_1 = np.asarray([11]*n)

    mean_class_2_domain_1  = [10, 10, 0, 10, 10, 0, 10, 10, 0]
    cov_class_2_domain_1  = np.diag([9,7,6,4,2,1,8,8,4])
    X_data_class_2_domain_1  = np.random.multivariate_normal(mean_class_2_domain_1, cov_class_2_domain_1, n)
    y_data_class_2_domain_1 = np.asarray([21]*n)

    X_data = np.concatenate((X_data_class_1_domain_1, X_data_class_2_domain_1), axis = 0)
    y_data = np.concatenate((y_data_class_1_domain_1, y_data_class_2_domain_1), axis = 0)
    n_samples = np.shape(X_data)[0]
    
    x_data = torch.from_numpy(X_data)
    y_data = torch.from_numpy(y_data)
    
    return n_samples, x_data, y_data

def create_target_data(n):

    mean_class_1_domain_2  = [10, 0, -10, 10, 0, -10, 10, 0, -10]
    cov_class_1_domain_2  = np.diag([8,4,3,0,2,3,1,3,3])
    X_data_class_1_domain_2  = np.random.multivariate_normal(mean_class_1_domain_2, cov_class_1_domain_2, n)
    y_data_class_1_domain_2 = np.asarray([12]*n)

    mean_class_2_domain_2  = [10, 10, 10, 10, 10, 10, 10, 10, 10]
    cov_class_2_domain_2  = np.diag([3,4,6,4,2,3,3,8,1])
    X_data_class_2_domain_2  = np.random.multivariate_normal(mean_class_2_domain_2, cov_class_2_domain_2, n)
    y_data_class_2_domain_2 = np.asarray([22]*n)

    X_data = np.concatenate((X_data_class_1_domain_2, X_data_class_2_domain_2), axis = 0)
    y_data = np.concatenate((y_data_class_1_domain_2, y_data_class_2_domain_2), axis = 0)
    n_samples = np.shape(X_data)[0]
    
    x_data = torch.from_numpy(X_data)
    y_data = torch.from_numpy(y_data)
    
    return n_samples, x_data, y_data
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset

from torch.utils.tensorboard import SummaryWriter
```

```python
class CNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(CNN, self).__init__()
        
        """
        formula [(WK+2P)/S]+1.
        """
        self.conv1 = nn.Conv1d(input_size, 64, kernel_size=2, stride=1)#input: 9 #18000
        self.conv2 = nn.Conv1d(64,32,kernel_size=2, stride = 1, padding=1)#input: [(9-2+2*0)/1]+1 = 8
        self.batch1 =nn.BatchNorm1d(32)#input: [(925-10+2*1)/1]+1 = 918 #[(8-2+2*1)/1]+1 = 9 
        self.conv3 = nn.Conv1d(32,32,kernel_size=1, stride = 1, padding=1) #input:9
        self.batch2 =nn.BatchNorm1d(32)#input: [(9-1+2*1)/1]+1 = 11
        #self.fc1 = nn.Linear(32*916, output_size)

    def forward(self, x):
        x = F.selu(self.conv1(x)) #conv1
        x = self.conv2(x) #conv2
        x = F.selu(self.batch1(x)) #batch1
        x = self.conv3(x) #conv3
        x = F.selu(self.batch2(x)) #batch2
        x = torch.reshape(x,(x.shape[0],x.shape[1]*x.shape[2])) #flatten
        #x = self.fc1(x) #linear1
        output = x
        
        return output
```

```python
def rbf_mmd2(Xs, Xt, sigma):
    
    xx, yy, zz = torch.mm(Xs, Xs.t()), torch.mm(Xt, Xt.t()), torch.mm(Xs, Xt.t())
    rx = (xx.diag().unsqueeze(0).expand_as(xx))
    ry = (yy.diag().unsqueeze(0).expand_as(yy))
    
    dxx = rx.t() + rx - 2. * xx # Used for A in (1)
    dyy = ry.t() + ry - 2. * yy # Used for B in (1)
    dxy = rx.t() + ry - 2. * zz # Used for C in (1)
    
    XX, YY, XY = (torch.zeros(xx.shape),
                  torch.zeros(xx.shape),
                  torch.zeros(xx.shape))

    for a in sigma:
        XX += torch.exp(-0.5*dxx/a)
        YY += torch.exp(-0.5*dyy/a)
        XY += torch.exp(-0.5*dxy/a)

    return torch.mean(XX + YY - 2. * XY)
```
