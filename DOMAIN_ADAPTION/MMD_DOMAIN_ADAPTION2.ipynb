```python
class CNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(CNN, self).__init__()
        
        """
        formula [(W−K+2P)/S]+1.
        """
        self.conv1 = nn.Conv1d(input_size, 64, kernel_size=100, stride=1)#input: 1024 #18000
        self.conv2 = nn.Conv1d(64,32,kernel_size=10, stride = 1, padding=1)#input: [(1024-100+2*0)/1]+1 = 925 #[(18000-100+2*0)/1]+1 = 17901
        self.batch1 =nn.BatchNorm1d(32)#input: [(925-10+2*1)/1]+1 = 918 #[(17901-10+2*1)/1]+1 = 17894 
        self.conv3 = nn.Conv1d(32,32,kernel_size=5, stride = 1, padding=1) #input:918 #17894
        self.batch2 =nn.BatchNorm1d(32)#input: [(918-5+2*1)/1]+1 = 916  #[(17894-5+2*1)/1]+1=17892
        #self.fc1 = nn.Linear(32*916, output_size)

    def forward(self, x):
        x = F.selu(self.conv1(x)) #conv1
        x = self.conv2(x) #conv2
        x = F.selu(self.batch1(x)) #batch1
        x = self.conv3(x) #conv3
        x = F.selu(self.batch2(x)) #batch2
        x = torch.reshape(x,(x.shape[0],x.shape[1]*x.shape[2])) #flatten
        #x = self.fc1(x) #linear1
        output = x
        
        return output
    
```

```python
#from mmd_loss import MMD_loss
class MMD_loss(nn.Module):
    def __init__(self, kernel_mul = 2.0, kernel_num = 5):
        super(MMD_loss, self).__init__()
        self.kernel_num = kernel_num
        self.kernel_mul = kernel_mul
        self.fix_sigma = None
        return
    def gaussian_kernel(self, source, target, kernel_mul=2.0, kernel_num=5, fix_sigma=None):
        n_samples = int(source.size()[0])+int(target.size()[0])
        total = torch.cat([source, target], dim=0)
        
        total0 = total.unsqueeze(0).expand(int(total.size(0)), int(total.size(0)), int(total.size(1)))
        total1 = total.unsqueeze(1).expand(int(total.size(0)), int(total.size(0)), int(total.size(1)))
        L2_distance = ((total0-total1)**2).sum(2) 
        if fix_sigma:
            bandwidth = fix_sigma
        else:
            bandwidth = torch.sum(L2_distance.data) / (n_samples**2-n_samples)
        bandwidth /= kernel_mul ** (kernel_num // 2)
        bandwidth_list = [bandwidth * (kernel_mul**i) for i in range(kernel_num)]
        kernel_val = [torch.exp(-L2_distance / bandwidth_temp) for bandwidth_temp in bandwidth_list]
        return sum(kernel_val)

    def forward(self, source, target):
        batch_size = int(source.size()[0])
        kernels = self.gaussian_kernel(source, target, kernel_mul=self.kernel_mul, kernel_num=self.kernel_num, fix_sigma=self.fix_sigma)
        XX = kernels[:batch_size, :batch_size]
        YY = kernels[batch_size:, batch_size:]
        XY = kernels[:batch_size, batch_size:]
        YX = kernels[batch_size:, :batch_size]
        loss = torch.mean(XX + YY - XY -YX)
        return loss
```

```python
def forward(model, classifier_layer_1, classifier_layer_2, windows, labels_source, labels_target, criterion, mmd_calculator, GAMMA_inter_class, GAMMA_intra_class, mmd_calculator_flag):
        
        #Feature Extraction
        outputs = model(windows.float())

        #Fully Connected 1 for mmd loss features
        x_src_1 = classifier_layer_1(outputs[:batch_size, :])
        x_tar_1 = classifier_layer_1(outputs[batch_size:, :])
        
        #Fully Connected 1 for classification
        x_src_2 = classifier_layer_2(x_src_1)
        x_tar_2 = classifier_layer_2(x_tar_1)
        
        #cross entropy loss
        ce_loss_source = criterion(x_src_2, labels_source) #train data
        ce_loss_target = criterion(x_tar_2, labels_target) #test data
        
        # mmd loss
        mmd_loss = 0
        shuffle_iter = 50
    
        
        
        if mmd_calculator_flag == True:
            
            #collect mmd_loss features from target and source for different classes
            class_0_source_x = None
            class_1_source_x = None
            class_2_source_x = None
            class_3_source_x = None

            source_x_list = [class_0_source_x, class_1_source_x, class_2_source_x, class_3_source_x]

            class_0_target_x = None
            class_1_target_x = None
            class_2_target_x = None
            class_3_target_x = None

            target_x_list = [class_0_target_x, class_1_target_x, class_2_target_x, class_3_target_x]

            for label_iter in range(x_src_2.size()[1]):
            
                source_x_list[label_iter] = x_src_1[labels_source==label_iter,:]
                target_x_list[label_iter] = x_tar_1[labels_target==label_iter,:]
            
            
            mmd_inter_class = 0
            mmd_intra_class = 0
        
            #calculate mmd loss between source and targert and between classes
            for source_iter in range(len(source_x_list)):
                for target_iter in range(len(target_x_list)):
                    
                    #get smallest length of feature vector to calculate mmd loss
                    smallest_lenth = min(source_x_list[source_iter].size()[0], target_x_list[target_iter].size()[0])
                    mmd_loss_value = mmd_calculator.forward(source_x_list[source_iter][:smallest_lenth,:], target_x_list[target_iter][:smallest_lenth,:])
                    
                    #mmd distance between same class should be minimized
                    if source_iter == target_iter:
                        mmd_inter_class += mmd_loss_value
                    
                    #mmd distance between different classes should be maximized
                    else:
                        mmd_intra_class += mmd_loss_value
            
            #weight inter class and intra class loss 
            mmd_loss = GAMMA_inter_class * mmd_inter_class - GAMMA_intra_class * mmd_intra_class
            
            #total loss  = mmd + ce
            loss = ce_loss_source + mmd_loss
            
        else:
            #total loss = ce 
            mmd_loss = 0
            loss = ce_loss_source
        

        #collect information about labels, predictions
        n_correct_source = 0
        n_correct_target = 0
        n_samples_source = 0
        n_samples_target = 0
        
        #for each element in batch check if prediction is correct and collect total and correct predictions and labels
        for i in range(len(labels_source)):
            label_source = labels_source[i]
            output_source = torch.argmax(x_src_2[i])
            if label_source == output_source:
                n_correct_source+=1
            n_samples_source+=1
        
        acc_total_source = 100.0 * n_correct_source / n_samples_source
            
        for i in range(len(labels_target)):
            label_target = labels_target[i]
            output_target = torch.argmax(x_tar_2[i])
            if label_target == output_target:
                n_correct_target+=1
            n_samples_target+=1

            

        acc_total_target = 100.0 * n_correct_target / n_samples_target
        
        return loss, mmd_loss, ce_loss_target, acc_total_source, acc_total_target
```

```python
writer_graph = SummaryWriter('runs/Dataloader2/graph')
writer_train = SummaryWriter('runs/Dataloader2/train')
writer_val = SummaryWriter('runs/Dataloader2/val')
writer_test = SummaryWriter('runs/Dataloader2/test')
writer = {}
writer["train"] = writer_train
writer["val"] = writer_val
writer["test"] = writer_test


#define training params
num_epochs = 50
learning_rate = 0.1#0.008
GAMMA_inter_class = 10
GAMMA_intra_class = 10
SIGMA = [2,4,8]  # default 1

#define loss and optimizer
criterion = torch.nn.CrossEntropyLoss()
mmd_calculator = MMD_loss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)


#dataloader for training and validation
dataloader_train = {}
dataloader_train["source"]=source_loader["train"]
dataloader_train["target"]=target_loader["test"]

#dataloader for testing
dataloader_val = {}
dataloader_val["source"]=source_loader["val"]
dataloader_val["target"]=target_loader["val"]

dataloaders = {}
dataloaders["train"] = dataloader_train
dataloaders["val"] = dataloader_val


#collect loss for each batch
ce_loss_collected = 0
mmd_loss_collected = 0
ce_loss_target_collected = 0
loss_list = {}
loss_list['train']=[]
loss_list['val']=[]

#collect mmd loss
mmd_loss_list=[]

#collect ce_loss_target
ce_loss_target_list = []


#collect accuracy for each batch
acc_total_source_collected = 0
acc_total_target_collected = 0
accuracy_list={}
accuracy_list['train']=[]
accuracy_list['val']=[]
accuracy_list['test']=[]


phases = ["train", "val"]
# Train and Validate the model
for epoch in range(num_epochs):
    loss_collected = 0
    acc_collected = 0
    mmd_loss_marker = True
    for phase in phases:
        iter_loader_source = iter(dataloaders[phase]["source"])
        iter_loader_target = iter(dataloaders[phase]["target"])
        
        mmd_calculator_flag = True
        for i in range(len(dataloaders[phase]["source"])):
            total_len_loader = len(dataloaders[phase]["source"])
            if i> 0.1*total_len_loader:
                mmd_calculator_flag = False
            ########Forward pass########
            windows_source, labels_source = iter_loader_source.next() #batch_size number of windows and labels from source domain
            windows_target, labels_target = iter_loader_target.next() #batch_size number of windows from target domain
            windows = torch.cat((windows_source, windows_target), dim=0) #concat the windows to 2*batch_size number of windows

            batch_size = len(labels_source) #take length of shorter dataoader which is the one from source domain (reason:train, val split)
                        
            if phase == "val":
                
                model.train(False) #no training
                
                with torch.no_grad():
                    loss, mmd_loss, ce_loss_target, acc_total_source, acc_total_target = forward(model, classifier_layer_1, classifier_layer_2, windows, labels_source, labels_target, criterion, mmd_calculator, GAMMA_inter_class, GAMMA_intra_class, mmd_calculator_flag)
                    
                    loss_collected += loss
                    mmd_loss_collected += mmd_loss
                    ce_loss_target_collected += ce_loss_target
                    acc_total_source_collected += acc_total_source
                    acc_total_target_collected += acc_total_target             
            
            elif phase == "train":
                
                model.train(True) # training
                
                                
                ########Forward pass########
                loss, mmd_loss, ce_loss_target, acc_total_source, acc_total_target = forward(model, classifier_layer_1, classifier_layer_2, windows, labels_source, labels_target, criterion, mmd_calculator, GAMMA_inter_class, GAMMA_intra_class, mmd_calculator_flag)
                
                loss_collected += loss
                mmd_loss_collected += mmd_loss
                ce_loss_target_collected += ce_loss_target
                acc_total_source_collected += acc_total_source
                acc_total_target_collected += acc_total_target  

                ########Backward pass########
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
    
        running_loss = loss_collected / len(dataloaders[phase]["source"])
        running_acc_source = acc_total_source_collected / len(dataloaders[phase]["source"])
        
        running_ce_loss_target = ce_loss_target_collected / len(dataloaders[phase]["source"])
        running_acc_target = acc_total_target_collected / len(dataloaders[phase]["source"])
        
        running_mmd_loss = mmd_loss_collected/ len(dataloaders[phase]["source"])
        
        loss_collected = 0
        mmd_loss_collected = 0
        ce_loss_target_collected = 0
        acc_total_source_collected = 0
        acc_total_target_collected = 0
        
        #list of train and validation loss for later plot
        if phase == "train":
            #test loss and accuracy tensorboard
            writer["test"].add_scalar(f'test_loss', running_ce_loss_target, epoch)
            ce_loss_target_list.append(running_ce_loss_target)
            
            writer["test"].add_scalar(f'mmd_loss', running_mmd_loss, epoch)
            mmd_loss_list.append(running_mmd_loss)
            
            writer["test"].add_scalar(f'accuracy', running_acc_target, epoch)
            accuracy_list["test"].append(running_acc_target)
            
            loss_list[phase].append(running_loss.detach())
            
        elif phase == "val":
            loss_list[phase].append(running_loss)
            
        #list of train and validation accuracy for later plot
        accuracy_list[phase].append(running_acc_source)
        
        #train and validation loss and accuracy tensorboard
        writer[phase].add_scalar(f'training loss', running_loss, epoch)
        writer[phase].add_scalar(f'accuracy', running_acc_source, epoch)
        
        



            

    print(f"Epoch {epoch+1}/{num_epochs} successfull")

```
